```python
#a) Series:

#- A one-dimensional array-like structure with homogeneous data

#- Can be thought of as a single column of data

#- Index-value pairs, where each value has a corresponding index

#- Can be created from:

 # * Lists: pd.Series([1, 2, 3, 4])

    #* Dictionaries: pd.Series({'a': 1, 'b': 2, 'c': 3})

    #* Scalar values: pd.Series(5, index=['a', 'b', 'c'])

    #- Cannot be created from images

```
